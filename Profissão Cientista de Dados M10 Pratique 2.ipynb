## Análise extra

````python
ordem_educacao = [
    "Primário", 
    "Secundário", 
    "Superior incompleto", 
    "Superior completo", 
    "Pós graduação"
]








plt.figure(figsize=(10, 3))
plot_sexo = sns.pointplot(df, x = "educacao", y = "renda", hue="sexo", order=ordem_educacao)

plt.xlabel("Educação")
plt.ylabel("renda")
plt.grid(True, axis='y', linestyle='--', alpha=0.3)
plt.show()
````

````python
plot_sexo = sns.pointplot(df, x = "sexo", y = "renda")
````

````python
#tirar outlier como nas aulas anteriores para ver se melhora
df["log_renda"] = np.log1p(df["renda"])

plot_sexo = sns.pointplot(df, x = "sexo", y = "log_renda")
````

 Mesmo fazendo o Log para diminuir o impacto de possiveis Outliers, homens continuam recebendo uma média bem acima das mulheres
